```python
import sys
import pickle
```

```python
class Employee:
    '''
        The __init()__ method of the Employee class accepts parameters for all instance variables specific for an employee.
    ''' 
    def __init__(self, emp_type, emp_name , emp_add, veh_data):
        self.__type = emp_type
        self.__employee_name = emp_name
        self.__employee_address = emp_add
        self.__employee_vehicle_info = veh_data
        
 #Define the accessor and mutator methods        

    def get_type(self):
        return self.__type
    
    def set_type(self, emp_type):
        self.__type = emp_type

    def get_name(self):
        return self.__employee_name
    
    def get_address(self):
        return self.__employee_address
    
    def get_vehicle_data(self):
        return self.__employee_vehicle_info 

    def set_name(self, emp_name):
        self.__employee_name = emp_name
        
    def set_address(self, emp_add):
        self.__employee_address = emp_add
        
    def set_vehicle_data(self):
        self.__employee_vehicle_info = vehicle_data
            
    def compute_compensation(self):
        pass
    
    '''
        Redefine the __str__() method and return a string that contains the values
        of the two instance variables of the Employee object
    '''
        
    def __str__(self):
        return "Employee Name: " + self.__employee_name + "; Employee address: " + self.__employee_address
```

```python
class FullTimeEmployee(Employee):
    '''
      The __init()__ method of the FullTimeEmployee class accepts parameters for all instance variables specific for a 
      full time employee and also inherits the variables from Employee class.
    '''
    
    def __init__(self, emp_type, emp_name, emp_add, vehicle_data, emp_salary):
        super().__init__(emp_type, emp_name, emp_add, vehicle_data)
        self.__employee_salary = emp_salary
        
    #Define the accessor and mutator methods
    def get_emp_salary(self):
        return self.__employee_salary 

    def set_emp_salary(self, emp_name):
        self.__employee_salary = emp_salary     
        
    #Full-Time Employee compensation calculation
    def compute_compensation(self):
        sal = self.get_emp_salary()
        if (sal <= 40000):
            tax = sal * 0.16
            return (sal - tax)
        elif (sal > 40000 and sal <= 75000):
            tax1 = 40000 * 0.16
            tax2 = (sal-40000) * 0.26
            return (sal - (tax1 + tax2))  
        elif (sal > 75000):
            tax1 = 40000 * 0.16
            tax2 = (75000-40000) * 0.26
            tax3 = (sal - 75000) * 0.33
            return (sal-(tax1 + tax2 + tax3))
              
    def __str__(self):
        allData = super().__str__()
        allData = "\nDetails of Full Time Employee are:" + allData
        allData += "Employee Salary: " + self.__employee_salary
        return allData
```

```python
class HourlyEmployee(Employee):
    '''
      The __init()__ method of the HourlyEmployee class accepts parameters for all instance variables specific for a 
      hourly employee and also inherits the variables from Employee class.
   '''
    def __init__(self, emp_type, emp_name, emp_add, vehicle_data, hrsWrkd, hrRate):
        super().__init__(emp_type, emp_name, emp_add, vehicle_data)
        self.__hoursWorked = hrsWrkd
        self.__hourlyRate = hrRate
        
    def getHoursWorked(self):
        return self.__hoursWorked
    
    def setHoursWorked(self, hrsWrkd):
        self.__hoursWorked = hrsWrkd
        
    def getHourlyRate(self):
        return self.__hourlyRate
    
    def setHourlyRate(self, hrRate):
        self.__hourlyRate = hrRate
        
    def compute_compensation(self):
        hw = self.getHoursWorked()
        hr = self.getHourlyRate()
        if(hw <= 40):
            return hw * hr
        else:
            return 40 * hr + (hw -40) * hr * 1.7
    
    def __str__(self):
        allData = super().__str__()
        allData = "\nDetails of Hourly Employee: "+allData
        allData += "\nHours worked: " + self.__hoursWorked+"Hourly Rate: "+self.__hourlyRate 
        return allData
```

```python
class Consultant(Employee):
    '''
      The __init()__ method of the Consultant class accepts parameters for all instance variables specific for a 
      consultant and also inherits the variables from Employee class.
   '''
    
    def __init__(self, emp_type, emp_name, emp_add, vehicle_data, hrsWrkd, prjTyp):
        super().__init__(emp_type, emp_name, emp_add, vehicle_data)
        self.__hoursWorked = hrsWrkd
        self.__projectType = prjTyp
        
#Define the accessor and mutator methods    

    def getHoursWorked(self):
        return self.__hoursWorked
    
    def setHoursWorked(self, hrsWrkd):
        self.__hoursWorked = hrsWrkd
        
    def getProjectType(self):
        return self.__projectType
    
    def setProjectType(self, prjTyp):
        self.__projectType = prjTyp
        
    def compute_compensation(self):
        hw = self.getHoursWorked()
        pt = self.getProjectType()
        if pt== 1:
            return (hw * 55)
        elif pt == 2:
            return (hw * 70)
        elif pt == 3:
            return (hw * 85)
        
    def __str__(self):
        allData = super().__str__()
        allData = "\nDetails of Consultant: "+allData
        allData += "\nHours worked: " + self.__hoursWorked+" Project type: "+self.__projectType 
        return allData
```

```python
class Vehicle():
    
    '''
      The __init()__ method of the Vehicle class accepts parameters for all instance variables specific for the 
      vehicle and all the variables are linked to the Employee by using aggregation.
   '''
    def __init__(self, make, model, year, mileage):
        self.__car_make = make
        self.__car_model = model
        self.__car_year = year
        self.__car_mileage = mileage 

    def get_car_make(self):
        return self.__car_make
    
    def set_car_make(self, make):
        self.__car_make = make
        
    def get_car_model(self):
        return self.__car_model
    
    def set_car_model(self, model):
        self.__car_model = model
        
    def get_car_year(self):
        return self.__car_year
    
    def set_car_year(self, year):
        self.__car_year = year
        
    def get_car_mileage(self):
        return self.__car_mileage
    
    def set_car_mileage(self, mileage):
        self.__car_mileage = mileage 
    
    def __str__(self):
        return "Make:"+self.__car_make+ "Model:"+self.__car_model+ "Year:"+self.__car_year+ "Mileage:"+self.__car_mileage
```
