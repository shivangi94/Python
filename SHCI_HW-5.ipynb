<h2>Lambda Functions - Question One </h2>
Write a Python program that takes a list of strings and prints all the strings with length greater than or equal to 5.  You must use a Lambda function and the filter function to answer this question. Below is a sample interaction.


<h2>Sample Interaction for Lambda Functions - Question One
    
</h2>



````python
input_string = input('Enter the strings separated by comma ')
my_list = input_string.split(",")
#print('list: ', my_list)
results = filter(lambda x : True if len(x)>=5 else False, my_list)
print("The strings with more than 4 characters are: ")
for i in results:
    print(i)
````

<h2>Lambda Functions - Question Two </h2>
Write a Python program that does the following:

    1.  accepts a list of strings 
    2.  converts the string to upper case if the length of the string is greater than 3 characters
    3.  Makes no change to the string if its length is 3 characters or less
    4.  Prints out all the strings
    
    You must use a Lambda function and the map function to answer this question. Below is a sample interaction.
    
 

<h2> Sample Interaction for Lambda Functions - Question Two</h2>



````python
input_string = input('Enter the strings separated by comma ')
my_list = input_string.split(",")
#print('list: ', my_list)
results = map(lambda x : x.upper() if len(x)>3 else False, my_list)
print("The strings with more than 3 characters are converted to upper case: ")
for i in results:
    print(i)
````

<h2>Plots - Question One </h2>

The attached `carprices.txt` contains a single column of car prices in 1000’s of dollars.  Read this data into your program.  The first line is the heading and should be ignored.  For any line containing non-numeric data, raise a ValueError exception and print a message saying, ‘This data is invalid and will be ignored’.  For the remainder of the data, you should divide each number by 10 and round it down to the nearest integer.  Then plot a histogram that looks like the one shown in the diagram below.  The number of bins should be chosen so as to produce the graph shown below.

<h2> Output for Plots-Question One</h2>

>

````python
import matplotlib.pyplot as plt
import math
import csv

price = []

with open('carprices.txt', 'r') as csv_in:
    csv_reader = csv.DictReader(csv_in)
    for row in csv_reader:
        try:
            value=math.floor(int(row['Min Price'])/10)
            price.append((value))
            continue
        except ValueError:
            print("This data is invalid and will be igorned")

#print(price)            
range = (0,45) 
bins = 45

plt.hist(price, bins, range, color = 'cornflowerblue', histtype = 'bar', rwidth = 0.7) 

plt.xlabel('price in ten of thousands') 

plt.ylabel('No. of cars') 

plt.title('Frequencies of car prices') 

plt.show() 
````

<h2>Plots - Question Two </h2>

The attached `cartype.txt` contains a single columns listing the type of car (van, compact, etc).  Read this data into your program.  The first line is the heading and should be ignored.  Then plot a piechart after computing the frequency for each car type.  Your chart should look like the one shown below.  Each slice should be separated from the next as shown below.

<h2> Output for Plots-Question Two</h2>



````python
import matplotlib.pyplot as plt
f = open('cartype.txt', 'r')
f.seek(6,0)
#print('position of the file pointer', f.tell(), '\n') 
text = f.read()

text = text.lower()
words = text.split()
words = [word.strip('.,!;()[]') for word in words]
words = [word.replace("'s", '') for word in words]

unique = []
for word in words:
    if word not in unique:
        unique.append(word)

#unique.sort()
#print(unique)
slices = []
for i in unique:
    freq=0
    for line in words:
        if i==line:
            freq +=1
    slices.append(freq)
#print(slices)
            
colors = ['y', 'b', 'c'] 

plt.pie(slices, labels = unique, colors=colors, explode = (0.03, 0.03, 0.03, 0.03, 0.03, 0.03), 
        startangle=90, 
        radius = 1.2, autopct = '%2.2f%%') 

plt.show() 

f.close()
````

<h2>Plots - Question Three</h2>

The attached `wildlife.txt` contains data about the population of whales, dolphins and whales over a number of years. Read this data into your program.  The first line is the heading and should be ignored.  Then plot line charts displaying the population trend for each animal.  Your graph should look like the one shown below.  

<h2> Output for Plots-Question Three</h2>



````python
import matplotlib.pyplot as plt
f = open('wildlife.txt')
text_contents  = []
years = []
dolphins= []
whales = []
bears = []
text = f.readlines()
for t in text:
    text_contents.append(t.split())

for i in text_contents:
    if i[0].isdigit():
        years.append(int(i[0]))
        bears.append(int(i[1]))
        dolphins.append(int(i[2]))
        whales.append(int(i[3]))
    else:
        continue
        
plt.plot(years, bears, label = "Bears")
plt.plot(years, dolphins, label = "Dolphins")
plt.plot(years, whales, label = "whales")
plt.xlabel('Years') 

plt.ylabel('Wildlife Population')

plt.title('Wildlife Population per year')
  
plt.legend(loc='upper left') 

plt.show() 
f.close()
````
